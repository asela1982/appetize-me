```python
import string
import nltk

def create_ingredients_vector(ingredients_list, current_ingredients):
    """ Given a list of words to search for in 'ingredients' field, return
        one-hot-encoded vector for that field """  
    # Initialize our hot_encoder
    hot_encoder = [0] * len(ingredients_list)

    # Binary vector variable is initialized to zeros for all words in passed "ingredients_list" list
    binary_vector = dict(zip(ingredients_list, hot_encoder))

    # Build ingredient words into a list
    words = [ingr_word 
                 for ingredient in current_ingredients for ingr_word in ingredient.split()
            ]
    
    # Strip all punctuation out of ingredient words (retains "/" and "-")
    punctuation = {'[', '!', '_', '@', '(', '=', '<', '}', '\\', '&', '|', ')', ']', '?', '`', 
                   '{', '.', '"', ',', ':', '+', '#', '>', ';', '*', '$', '~', '%', "'", '^'}
    words_clean = []
    for word in words:
        words_clean.append(''.join(ch for ch in word if ch not in punctuation))
    
    # Remove plural from ingredient words
    ps = nltk.PorterStemmer()
    words_clean_singular = [ps.stem(word) for word in words_clean]
    print(words_clean_singular)
    
    # Loop through words in ingredients, and vectorize based on unique ingredients represented in our master list
    # (Assign a "1" to the ingredient vector word, if found in list.)
    for word in words_clean_singular:
        if word in ingredients_list and binary_vector[word] == 0:
            binary_vector[word] = 1
       
    return binary_vector
```

```python
def create_tags_vector(tags_list, current_tags):
    """ Given a list of words to search for in 'tags' field, return
        one-hot-encoded vector for that field """  
    # Initialize our hot_encoder
    hot_encoder = [0] * len(tags_list)

    # Binary vector variable is initialized to zeros for all words in passed "ingredients_list" list
    binary_vector = dict(zip(tags_list, hot_encoder))

    # Create list from tags in current row's value
    words = [tag for tag in current_tags]
    
    # Loop through words in tags, and vectorize based on unique tags represented in our master list
    for word in words:
        if word in tags_list and binary_vector[word] == 0:
            binary_vector[word] = 1
       
    return binary_vector
```
